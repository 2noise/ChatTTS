```python
# openai api()
from openai import OpenAI
from IPython.display import Audio, display

# 
client = OpenAI(
    api_key="dummy-key",
    base_url="http://localhost:8000/v1"
)

# 
response = client.audio.speech.create(
    model="tts-1",
    voice="echo",
    input= """ 
    1. Good morning! Wish you a wonderful day. 
    2. Hello there, how have you been recently? 
    3. Don't give up, you can do it! 
    4. Keep up the good work, your efforts will pay off.""",
    response_format="wav"
)

# 
audio_data = response.content  # response.content  bytes 

#  Notebook 
display(Audio(audio_data, autoplay=False))
```

```python
# request
import requests
from IPython.display import Audio, display
import io

payload = {
    "model": "tts-1",
    "input": """ 
    1. Good morning! Wish you a wonderful day. 
    2. Hello there, how have you been recently? 
    3. Don't give up, you can do it! 
    4. Keep up the good work, your efforts will pay off.""",
    "voice": "echo",
    "response_format": "wav", 
    "stream": True
}

try:
    response = requests.post("http://localhost:8000/v1/audio/speech", json=payload, stream=True)
    response.raise_for_status()  # 
    
    audio_buffer = io.BytesIO()
    for chunk in response.iter_content(chunk_size=8192):
        if chunk:
            audio_buffer.write(chunk)
    
    audio_buffer.seek(0)
    display(Audio(audio_buffer.getvalue(), autoplay=False))
    print(" Notebook")
except requests.exceptions.RequestException as e:
    print(f": {str(e)}")
    if hasattr(e.response, "text"):
        print(f": {e.response.text}")
```

```python
import subprocess

# ,wav
cmd = (
    'curl -X POST "http://localhost:8000/v1/audio/speech" '
    '-H "Content-Type: application/json" '
    '-d \'{"model": "tts-1", "input": " 1. Good morning! Wish you a wonderful day. 2. Hello there, how have you been recently? 3. Dont give up, you can do it! 4. Keep up the good work, your efforts will pay off.", "voice": "echo", "response_format": "wav", "stream": true}\' '
    '-s | mpv --no-video -'
)
subprocess.run(cmd, shell=True, check=True)
```
